```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas
import seaborn
iris_df = seaborn.load_dataset('iris')
```

```python
iris_df.head()
```

```python
plt.scatter(iris_df.sepal_length, iris_df.sepal_width)
```

```python
plt.scatter(iris_df.sepal_length, iris_df.sepal_width)
ax = plt.gca()
for i, tick in enumerate(ax.get_xticklabels()):
    tick.set_rotation(i * 20)
    tick.set_fontsize(i * 5)
```

```python
plt.scatter(iris_df.sepal_length, iris_df.sepal_width)
ax = plt.gca()
for tick in ax.get_xticklabels():
    tick.set_rotation(45)
```

```python
p1 = plt.subplot(2,1,1)
p2 = plt.subplot(2,1,2)
p1.scatter(iris_df.sepal_length, iris_df.sepal_width)
p2.scatter(iris_df.petal_length, iris_df.petal_width)
p1.set_xlabel('sepals')
p2.set_xlabel('petals')
for ax in (p1, p2):
    for i, tick in enumerate(ax.get_xticklabels()):
        tick.set_rotation(i * 20)
    
plt.tight_layout()
```

```python
plt.subplot(2,1,1)
plt.scatter(iris_df.sepal_length, iris_df.sepal_width)
plt.xlabel('sepals')

plt.subplot(2,1,2)
plt.scatter(iris_df.petal_length, iris_df.petal_width)
plt.xlabel('petals')
plt.tight_layout()
```

```python
plt.scatter(iris_df.sepal_length, iris_df.sepal_width)
ax = plt.gca()
for _, data in iris_df.iterrows():
    if data.sepal_length < 6.0:
        ax.annotate(data.species, (data.sepal_length + 0.1, data.sepal_width))
for x, y, txt in zip(iris_df.sepal_length, iris_df.sepal_width, iris_df.species):
    ax.annotate(txt, (x + 0.1, y))
```

```python
ax = iris_df.plot.scatter(x='sepal_length', y='sepal_width')
for x, y, txt in zip(iris_df.sepal_length, iris_df.sepal_width, iris_df.species):
    ax.annotate(txt, (x + 0.1, y))
```

```python
r = seaborn.jointplot(x='sepal_length', y='sepal_width', data=iris_df)
for x, y, txt in zip(iris_df.sepal_length, iris_df.sepal_width, iris_df.species):
    r.ax_joint.annotate(txt, (x + 0.1, y))
```

```python
seaborn.distplot(iris_df.sepal_width, kde=False)
```

```python
with plt.xkcd():
    iris_df.plot.hist(bins=50)
```

```python
plt.style.use('ggplot')
iris_df.plot.hist(bins=50)
```

```python
# plt.style.use('bmh')
seaborn.kdeplot(iris_df.sepal_length)
seaborn.kdeplot(iris_df.sepal_width)
seaborn.kdeplot(iris_df.petal_length)
seaborn.kdeplot(iris_df.petal_width)
```

```python
seaborn.scatterplot(x=iris_df.sepal_length, y=iris_df.sepal_width, hue=iris_df.species)
```

```python
color_map = {'setosa': 'green', 'versicolor': 'red', 'virginica': 'blue'}
colors = [color_map[x] for x in iris_df.species]
print(colors)
iris_df.plot.scatter(x='sepal_length', y='sepal_width', c=colors)
```
